```python
#Libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv(r"F:\Downloads\Dataset_spine.csv")
```

```python
#Dataset analysis
print("The first 5 rows :\n ",df.head(5))
print("\n")
print("The last 5 rows : \n",df.tail(5))
```

```python
print("Shape of dataset : ", df.shape)
```

```python
#statistical data
df.describe()
```

```python
#Dataset cleansing
df = df.drop(['Unnamed: 13'], axis=1)
```

```python
df = df.drop(['Col7','Col8','Col9','Col10','Col11','Col12'], axis=1)
```

```python
df.head()
```

```python
#Defining values of x and y respectively
y = df['Class_att']
x = df.drop(['Class_att'], axis=1)
```

```python
# Train test split - 10%

#Transform the data in training and testing
#train_test_split selects randomly the train and test size basing on the ratio given. 
#Every single time you run this function you will have a randomly selected train and test values
#based on the train and test size ratio. 
#This random selection every particular time you run this results in the "random_states". 
#To avoid getting different values for train and test every time you select the random_state you want to use. 
#This will also have an impact on the evaluation metrics 
#as if the random_state is not selected the values from the evaluation will differ.
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.10, random_state=27)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(100,100,100), max_iter=500, alpha=0.0001,
                     solver='sgd', verbose=10,  random_state=21,tol=0.000000001)
```

```python
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
#testing

accuracy_score(y_test, y_pred)
```

```python
#A confusion matrix is a table that is often used to describe the performance of a classification model (or "classifier")
from sklearn.metrics import classification_report,confusion_matrix
print("Classification report \n=======================")
print(classification_report(y_true=y_test, y_pred=y_pred))
print("Confusion matrix without normalization\n=======================")
print(confusion_matrix(y_true=y_test, y_pred=y_pred))
```

```python
test_size = [.10,.20,.30,.40,.50]
acc = {.10:0,.20:0,.30:0,.40:0,.50:0}
```

```python
for t in test_size :
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= t, random_state=27)
    clf = MLPClassifier(hidden_layer_sizes=(100,100,100), max_iter=500, alpha=0.0001,
                     solver='sgd', verbose=10,  random_state=21,tol=0.000000001)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    acc[t] = accuracy
```

```python
print("The accuracies for different values of train-test split is \n\n", acc)
```
